```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 
```

# Visualization Drill

```python
df = pd.read_csv('movie_ratings.csv')
```

```python
df.head()
```

```python
# Renaming Columns
df.rename(columns={'Day of Week':'Day_of_Week'}, inplace = True) 
df.rename(columns={'Adjusted Gross ($mill)':'Adjusted_Gross'}, inplace = True) 
df.rename(columns={'Budget ($mill)':'Budget'}, inplace = True) 
df.rename(columns={'Gross ($mill)':'Gross'}, inplace = True) 
df.rename(columns={'IMDb Rating':'IMDb_Rating'}, inplace = True)
df.rename(columns={'MovieLens Rating':'MovieLens_Rating'}, inplace = True) 
df.rename(columns={'Overseas ($mill)':'Overseas'}, inplace = True)
df.rename(columns={'Profit ($mill)':'Profit'}, inplace = True) 
df.rename(columns={'US ($mill)':'US'}, inplace = True) 
df.rename(columns={'Movie Title':'Movie'}, inplace = True) 
df.rename(columns={'Release Date':'Release_Date'}, inplace = True) 
df.rename(columns={'Runtime (min)':'Runtime'}, inplace=True)
```


    Choose one variable and plot that variable four different ways.


```python

sns.set(style='whitegrid')

plt.hist(df.US)
plt.suptitle('US Gross of Films')
plt.show()

sns.kdeplot(df.US)
plt.xlabel('US Total Gross')
plt.ylabel('Frequency')
plt.title('US Gross of Films Frequency Distribution');
plt.show()


plt.title('US Gross of Films Boxplot');
ax = sns.boxplot(y='US', data=df)
plt.show()

plt.title('Genre of Films');
sns.countplot(x="Genre", data=df)
plt.xticks(rotation = 90);


```

Choose two continuous variables, and plot them three different ways.

```python
g = sns.lmplot(y='Gross', 
               x='Budget',
               data=df,
               fit_reg=True,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Gross")
g.set_xlabels("Movie Budget")
plt.title('Scatterplot: Total Movie Gross by Budget')
plt.show()

plt.hist(df.Profit, color='red',  alpha=.5,label='Gross')
plt.hist(df.Budget, color='blue', alpha=.5,label='Budget')
plt.xlabel('Gross vs Budget')
plt.legend(loc='upper right')
plt.show()


sns.kdeplot(df.Profit, df.Budget, shade=True)
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.title('Gross / Budget Frequency');
plt.show()

sns.kdeplot(df.Gross)
sns.kdeplot(df.Budget)
plt.xlabel('Gross')
plt.ylabel('Frequency')
plt.title('Gross and Budget of Films Frequency Distribution');
plt.show()
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
monday = df.loc[(df['Day_of_Week']=='Monday'),'Gross']
tuesday = df.loc[(df['Day_of_Week']=='Tuesday'),'Gross']
wednesday = df.loc[(df['Day_of_Week']=='Wednesday'),'Gross']
thursday = df.loc[(df['Day_of_Week']=='Thursday'),'Gross']
friday = df.loc[(df['Day_of_Week']=='Friday'),'Gross']
saturday = df.loc[(df['Day_of_Week']=='Saturday'),'Gross']
sunday = df.loc[(df['Day_of_Week']=='Sunday'),'Gross']
```

```python
plt.hist(monday, color='red',  alpha=.5,label='Monday')  # alpha just controls the opacity
plt.hist(tuesday, color='blue', alpha=.5,label='Tuesday')
plt.hist(wednesday, color='green', alpha=.5,label='Wednesday')  
plt.hist(thursday, color='yellow', alpha=.5,label='Thursday')
plt.hist(friday, alpha=.5,label='Friday')
plt.hist(saturday, alpha=.5,label='Saturday')  
plt.hist(sunday, alpha=.5,label='Sunday')
plt.xlabel('Total Gross by Weekday')
plt.legend(loc='upper right')
plt.title('Plot 1: Total Gross by Weekday')
plt.show()
```
