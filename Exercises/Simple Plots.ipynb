```python
np.random.seed(1221)
```

```python
df = pd.DataFrame()
```

```python
df['rand'] = np.random.rand(100)
```

```python
df['rand_sq'] = df['rand'] ** 2
```

```python
df['rand_shift'] = df['rand'] + 2
```

```python
df['counts_sq'] = df.index ** 2
```

```python
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['rand'])
```

```python
plt.plot(df['rand'], color='purple')
```

```python
plt.ylim(-0.1,1.1)
```

```python
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```
